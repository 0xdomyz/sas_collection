```python
import saspy
sas = saspy.SASsession(cfgname='oda', colorLOG=True)

def run(code):
    sas.submitLST(code, method = 'logandlist')
```

```python
code = f"""
proc print data=sashelp.air (obs=5); 
run;
"""
run(code)
```

```python
run(f"""
data work.air;
    set sashelp.air;
    retain total 0;
    total + air;
run;

proc print data=work.air (obs=5);
"""
)
```

```python
sas.sasdata('air', "sashelp").columnInfo()
```

```python
sas.submitLST(f"""
data noformat_air;
    set sashelp.air;
    format _all_;
run;

proc print data=noformat_air (obs=5);
run;
""")
```

```python
run(f"""
data work.air;
    set sashelp.air;
    year = year(date);
run;

data yearly_totals;
    set air;
    by year;
    retain yearly_total 0;
    if first.year then yearly_total = 0;
    yearly_total + air;
run;

proc print data=yearly_totals;
run;
""")
```
