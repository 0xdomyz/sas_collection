```python
import saspy
sas = saspy.SASsession()
```

### arrays


```python
sas.sasdata("cars","sashelp").head()
```

```python
# * for auto array size, dim() for iteration number;
sas.submitLST(
f"""
data test;
    set sashelp.cars;

    array vars[*] mpg_city mpg_highway;
    array vars_out[*] mpg_city_k mpg_highway_k;

    do i=1 to dim(vars);
        vars_out[i] = vars[i] * 1000;
    end;
run;

proc print data=test (obs=5);
    var mpg_city mpg_city_k mpg_highway mpg_highway_k;
run;
""")
```

### auto vars

```python
sas.submitLST(
f"""
data test;
    set sashelp.cars;

    array vars[*] _numeric_;
    array chars[*] _character_;

    do i=1 to dim(vars);
        vars[i] = vars[i] * 1000;
    end;

    do i=1 to dim(chars);
        chars[i] = cats(chars[i], '_k');
    end;

    drop i;
run;

proc print data=test (obs=5);
run;
""")
```
