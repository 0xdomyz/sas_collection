```python
import saspy
sas = saspy.SASsession()
```

### procedures

```python
# no required options varies
sas.submitLST(f"""
    data class;
        set sashelp.class;
    quit;
              
    proc print;
    run;
""")
```

many options/statements e.g.:
- options: data, out
- statements: 
    - by, where, output
    - title, footnote, label

```python
sas.submitLST(f"""
    proc sort data=sashelp.class out=class;
        by sex;
    run;

    proc print data=work.class;
        by sex;
        where age > 10;
        title 'Class Data';
        footnote 'there are 2 panels to this report';
        label name = 'Student Name';
        /* output out=work.class_res;*/
    run;

    footnote;
""")

```

### where statement

```python
sas.submitLST(f"""
data artists;
    input name $ 1-21 genre $ 23-42 origin $ 44-63;
    datalines;
Mary Cassatt           Impressionism       U
Paul Cezanne           Post-impressionism  F
Edgar Degas            Impressionism       F
Paul Gauguin           Post-impressionism  F
Claude Monet           Impressionism       F
Pierre Auguste Renoir  Impressionism       F
Vincent van Gogh       Post-impressionism  N
;
run;
              
proc print data=artists;
    where lowcase(genre) not contains 'post' | origin ~= 'F';
run;
""")
```
