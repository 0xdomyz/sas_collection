```python
import saspy
sas = saspy.SASsession()
```

### examples of using sysfunc and funcs to create macro variables

```python
sas.submitLST(
f"""
title "%sysfunc(date(),worddate.) Absence Report";
proc print data=sashelp.class (obs=1);
run;
title;
""")
```

```python
sas.submitLST(
f"""
%let string1 = V01N01-V01N10;
%let string2 = %sysfunc(translate(&string1,P, N));

%put &string2;

""", method="listandlog")
```

```python
# alternatively use dummy data step, and call symputx
sas.submitLST(
f"""
data _null_;
    string1 = "V01N01-V01N10";
    string2 = translate(string1, "P", "N");
    call symputx('string2', string2);
run;

%put &string2;
""", method="listandlog")
```
