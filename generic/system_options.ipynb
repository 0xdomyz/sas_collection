```python
import saspy
sas = saspy.SASsession()
```

### SOURCE2: Controls the display of SAS statements generated by macro execution and DATA step compilation.

```python
sas.submitLST(
f"""
options nosource2;

%macro mac(input);
    data _null_;
        put "Hello &input.!";
    run;
%mend mac;

%mac(input=World);

""", method="logandlist")
```

```python
sas.submitLST(
f"""
options source2;

%macro mac(input);
    data _null_;
        put "Hello &input.!";
    run;
%mend mac;

%mac(input=World);

options nosource2;
""", method="logandlist")
```
