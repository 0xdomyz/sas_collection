```python
import saspy
sas = saspy.SASsession()
```

```python
# reg
sas.submitLST(
f"""
proc sgplot data=sashelp.iris;
    reg x=SepalLength y=SepalWidth / group=Species;
run;
""")
```

```python
# reg
# prediction limits for individual predicted values
sas.submitLST(
f"""
proc sgplot data=sashelp.iris;
    reg x=SepalLength y=SepalWidth / group=Species cli;
    where Species = 'Setosa';
run;
""")
```

```python
# reg
# confidence limits for mean predicted values
sas.submitLST(
f"""
proc sgplot data=sashelp.iris;
    reg x=SepalLength y=SepalWidth / group=Species clm;
    where Species = 'Setosa';
run;
""")
```

```python
# loess
sas.submitLST(
f"""
proc sgplot data=sashelp.iris;
    loess x=SepalLength y=SepalWidth / group=Species;
run;
""")
```

```python
# pbspline
sas.submitLST(
f"""
proc sgplot data=sashelp.iris;
    pbspline x=SepalLength y=SepalWidth / group=Species;
run;
""")
```

```python
# reg and loess with conf intervals
# nomarkers to avoid double plotting
sas.submitLST(
f"""
proc sgplot data=sashelp.iris;
    loess x=SepalLength y=SepalWidth / clm;
    reg x=SepalLength y=SepalWidth / nomarkers;
    where Species = 'Setosa';
run;
""")
```
