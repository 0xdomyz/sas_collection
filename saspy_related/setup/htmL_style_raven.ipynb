```python
import saspy
sas = saspy.SASsession(colorLOG=True)
```

```python
print(sas.HTML_Style)
```

```python
sas.submitLST(
f"""
proc print data=sashelp.cars (obs=5);
run;
""")
```

```python
sas.submitLST(
f"""
proc freq data=sashelp.cars;
    tables origin * type;
run;
""")
```

```python
sas.submitLST(
f"""
proc compare base=sashelp.cars compare=sashelp.cars;
run;
""")
```

```python
sas.submitLST(
f"""
proc contents data=sashelp.cars;
run;
""")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Simulate 1k numbers from a standard normal distribution
data = np.random.standard_normal(1000)

# Plot the data
plt.hist(data, bins=30)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Standard Normal Distribution')
plt.show()
```
