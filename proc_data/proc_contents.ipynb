```python
import saspy
sas = saspy.SASsession(cfgname='oda', )

def run(code):
    sas.submitLST(code, method = 'logandlist')
```

```python
libref = 'SASHELP'
tbl = "cars"
tbl_name = f"{libref}.{tbl}"
```

```python
cols = """
cylinders
drivetrain
enginesize
horsepower
length
mpg_city
mpg_highway
msrp
make
model
origin
type
weight
wheelbase
"""

cols = cols.strip().split('\n')
cols
```

```python
code = f"""
PROC CONTENTS DATA={tbl_name} OUT=CONTENTS;
RUN;
"""
run(code)
```

```python
sdf = sas.sasdata("contents")
df = sdf.to_df()
df = df[df['NAME'].str.lower().isin(cols)]
df
```
