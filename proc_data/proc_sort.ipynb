```python
import saspy
sas = saspy.SASsession()
```

### general

```python
sas.sasdata("class","sashelp").head()
```

```python
# descending keyword
sas.submitLST(
f"""
proc sort data=sashelp.class out=class;
    by sex descending name;

proc print data=class (obs=5);
run;
""", method="listorlog")
```

```python
# dedup and dupout option
sas.submitLST(f"""
proc sort data=sashelp.class out=df nodupkey dupout=df2;
    by sex age;

proc print data=df (obs=5);
run;

proc print data=df2;
run;

""")
```

### chars

```python
sas.submitLST(
f"""
data df;
    input char $;
    datalines;
abc
1
abcda
abcd1
.
abd
b
;
run;

proc print data=df;
run;
""", method="listonly")
```

```python
# missing le num le alpha
# comp from left to right, in alphabetical order
sas.submitLST(
f"""
proc sort data=df out=df2;
    by char;

proc print data=df2;
run;
""", method="listorlog")
```
