```python
import saspy
sas = saspy.SASsession(cfgname='oda', colorLOG=True)
```

```python
sas.sasdata("cars", libref= "sashelp").head()
```

```python
sas.submit(
'''
PROC TRANSPOSE DATA=sashelp.cars OUT=cars_transposed (DROP=_NAME_);
    BY make;
    VAR msrp;
RUN;
'''
)
sas.sasdata('cars_transposed').to_df()
```

```python
sas.submitLOG("""
data fishdata;
   infile datalines missover;
   input Location & $10. Date date7.
         Length1 Weight1 Length2 Weight2 Length3 Weight3
         Length4 Weight4;
   format date date7.;
   datalines;
Cole Pond   2JUN95 31 .25 32 .3  32 .25 33 .3
Cole Pond   3JUL95 33 .32 34 .41 37 .48 32 .28
Cole Pond   4AUG95 29 .23 30 .25 34 .47 32 .3
Eagle Lake  2JUN95 32 .35 32 .25 33 .30
Eagle Lake  3JUL95 30 .20 36 .45
Eagle Lake  4AUG95 33 .30 33 .28 34 .42
;
""")
sas.sasdata('fishdata').to_df()
```

```python
sas.submitLOG("""
proc transpose data=fishdata
     out=fishlength(rename=(col1=Measurement));
   var length1-length4;
   by location date;
run;
""")

sas.sasdata('fishlength').to_df()
```
