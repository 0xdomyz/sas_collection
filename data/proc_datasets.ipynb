```python
import saspy
sas = saspy.SASsession(colorLOG=True)
```

### dataset to check tables in lib

```python
sas.submitLST(f"""
    proc datasets dd=sashelp;
    quit;

""")
```

### vtable to check tables

```python
sas.submitLST(f"""
    proc sql;
        create table vtable as
        select memname 
        from sashelp.vtable;
        *where libname = 'WORK';
    quit;
""")
```

```python
df = sas.sasdata('vtable')
df
```
