```python
import saspy
sas = saspy.SASsession()
```

### data

```python
sas.sasdata("cars","sashelp").head()
```

```python
sas.sasdata("cars","sashelp").top("origin")
```

### tests

```python
sas.submitLST(
f"""
title;
PROC TTEST H0 = 50000 data = sashelp.cars; 
    VAR msrp; 
run;
""", method="listorlog")
```

```python
sas.submitLST(
f"""
title;
PROC TTEST data = sashelp.cars (where = (origin in ('Asia', 'USA'))); 
    class origin;
    VAR msrp; 
run;
""", method="listorlog")
```

```python
sas.submitLST(
f"""
title;
PROC TTEST data = sashelp.cars; 
    paired msrp * invoice; 
run;
""", method="listorlog")
```
