```python
import saspy
sas = saspy.SASsession()
```

```python
df = sas.sasdata("cars","sashelp").to_df()
df.head()
```

```python
# proc corr
sas.submitLST(
f"""
title;
proc corr data=sashelp.cars;
run;
""", method="listorlog")
```

```python
df.select_dtypes(include=['float64']).corr()
```

```python
# proc corr var with 2
sas.submitLST(
f"""
title;
proc corr data=sashelp.cars;
    var mpg_city mpg_highway horsepower enginesize msrp;
    with cylinders;
run;
""", method="listorlog")
```

```python
# pearson kendall spearman options
sas.submitLST(
f"""
title;
proc corr data=sashelp.cars pearson kendall spearman;
    var mpg_city mpg_highway horsepower enginesize msrp;
    with cylinders;
run;
""", method="listorlog")
```
